```python
import pandas as pd
from matplotlib import pyplot as plt
from pylab import rcParams
```

```python
plt.style.use('fivethirtyeight')
```

```python
DATASET_PATH = './creditcard.csv'
PREPARED_DATASET_PATH = './creditcard_prepared.csv'
```

```python
fd = pd.read_csv(DATASET_PATH, sep=',')
```

Создадим объект Series для диаграмм с натуральным и логарифмическим масштабом.

```python
class_fd = pd.Series(fd['Class'].value_counts(), index=set(fd['Class']))
```

```python
class_fd.plot(kind='bar', figsize=(4, 4))
plt.title('Натуральный масштаб')
plt.show()
```

```python
class_fd.plot(kind='bar', logy=True, figsize=(4, 4))
plt.title('Логарифмический масштаб')
plt.show()
```

Теперь создадим переменную для хранения значений признака V1 и на основе нее построим две гистограммы.

```python
v_first_fd = fd['V1']
v_first_fd.index = fd['Class']
```

```python
rcParams['figure.figsize'] = 12, 5
plt.hist(v_first_fd[0], bins=20, color='grey', density=True, alpha=0.5, ec='black', label='Class 0')
plt.hist(v_first_fd[1], bins=20, color='red', density=True, alpha=0.5, ec='darkred', label='Class 1')
plt.legend(loc='upper left')
plt.xlabel('Class')
plt.show()
```
