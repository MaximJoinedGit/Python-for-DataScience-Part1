```python
class DataPipeline:
    
    """Подготовка исходных данных"""
    
    def __init__(self):
        
        """Параметры класса"""
        self.rooms_mode = None
        self.square_median = None
        self.life_to_square_coef = None
        self.kitchen_to_square_coef = None
        self.healthcare_1 = None
        self.scaler = None
        
        
    def fit(self, df):
        
        """Сохранение статистик"""
        
        # Расчет статистик
        self.rooms_mode = df['Rooms'].mode()
        self.square_median = df['Square'].median()
        self.life_to_square_coef = (df['LifeSquare'] / df['Square']).median()
        self.kitchen_to_square_coef = (df['KitchenSquare'] / df['Square']).median()
        self.healthcare_1 = df['Healthcare_1'].median()
        self.scaler = StandardScaler()
    
    def transform(self, df):
        
        """Трансформация данных"""
        
        # 1. Заменим буквенные значения в выборке на цифренные.
        data_to_change_type = df.select_dtypes(include='object').keys()
        for column in data_to_change_type:
            old_values = np.sort((pd.unique(df[column])))
            new_values = [num for num in range(1, len(old_values) + 1)]
            for val in range(len(pd.unique(df[column]))):
                df[column] = np.where(df[column]==old_values[val], new_values[val], df[column])
            df[column] = df[column].astype(int)
        
        # 2. Колонка Rooms. Заменим все значения, которые меньше 1 и больше 5 на моду.
        df['Rooms'] = np.where(((df['Rooms'] < 1) | (df['Rooms'] > 5)), self.rooms_mode, df['Rooms'])
        
        # 3. Колонка Square. Заменим все значения, которые меньше 20 и больше 200 на среднее значение.
        df['Square'] = np.where(((df['Square'] < 20) | (df['Square'] > 200)), self.square_median, df['Square'])
        
        # 4. Колонка LifeSquare. Для начала посчитаем медиану отношения жилой площади к общей площади.
        # Теперь, с его помощью заполним пустые значения, значения меньше 20 и больше 200.
        df['LifeSquare'] = np.where((
            (df['LifeSquare'].isnull()) | (df['LifeSquare'] < (20 * self.life_to_square_coef)) | (df['LifeSquare'] > (200 * self.life_to_square_coef))), 
            (df['Square'] * self.life_to_square_coef), 
            df['LifeSquare'])
        
        # Также существует много значений, где сумма LifeSquare и KitchenSquare больше значения Square. 
        # Обработаем эти случаи, поменяв значения в LifeSquare, используя self.life_to_square_coef.
        df['LifeSquare'] = np.where(
            (df['Square'] < (df['LifeSquare'] + df['KitchenSquare'])),
            (df['Square'] * self.life_to_square_coef),
            df['LifeSquare'])
        
        # 5. Колонка KitchenSquare. Также считаем медиану отношения площади кухни к общей площади.
        # Заполним значения, которые меньше 20 и больше 200.        
        df['KitchenSquare'] = np.where((
            (df['KitchenSquare'] < (20 * self.kitchen_to_square_coef)) | (df['KitchenSquare'] > (200 * self.kitchen_to_square_coef))), 
            (df['Square'] * self.kitchen_to_square_coef), 
            df['KitchenSquare'])
        
        # Повторим действия из пункта 4, для изменения значений KitchenSquare, используя self.kitchen_to_square_coef.        
        df['KitchenSquare'] = np.where(
            (df['Square'] < (df['LifeSquare'] + df['KitchenSquare'])),
            (df['Square'] * self.kitchen_to_square_coef),
            df['KitchenSquare'])
        
        # 6. Колонка HouseFloor. Для начала обработаем выбросы.
        df['HouseFloor'] = np.where((df['HouseFloor'] > 60), 60, df['HouseFloor'])
        
        # Теперь обработаем данные, где этаж больше этажности. 
        # Для значений меньше 5 этажность заменим на 5, где меньше 10 на 9, где больше 10, на значение этажа.
        df['HouseFloor'] = np.where(((df['HouseFloor'] < df['Floor']) & (df['Floor'] <= 5)), 5, df['HouseFloor'])
        df['HouseFloor'] = np.where(((df['HouseFloor'] < df['Floor']) & (df['Floor'] <= 10)), 10, df['HouseFloor'])
        df['HouseFloor'] = np.where(((df['HouseFloor'] < df['Floor']) & (df['Floor'] <= 15)), 15, df['HouseFloor'])
        df['HouseFloor'] = np.where(((df['HouseFloor'] < df['Floor']) & (df['Floor'] <= 20)), 20, df['HouseFloor'])
        df['HouseFloor'] = np.where((df['HouseFloor'] < df['Floor']), df['Floor'], df['HouseFloor'])
        
        # 7. Колонка Healthcare_1. Заполним пустые значения средним значением.
        df['Healthcare_1'].fillna(0, inplace=True)
        
        # 8. Добавим новые фичи.
        
        # Средняя площадь комнат.
        df['RoomSquare'] = df['LifeSquare'] / df['Rooms']        
        
        # Из-за сильной корреляции признаков Social_1 и Social_2 используем их отношение.
        df['Social'] = df['Social_1'] / df['Social_2']
        
        # Коэффициент, показывающий, насколько далеко находится квартира от первого (0) и последнего (1) этажа.
        df['FloorRank'] = df['Floor'] / df['HouseFloor']
        
        # Коэффициент, определяющищй долю площади кухни от общей площади.
        df['KitchenCoef'] = df['KitchenSquare'] / df['Square']
        
        df['Social'] = df['Social_1'] / df['Social_2']

#         scaler = RobustScaler()
        
#         eco_scaled = pd.DataFrame(scaler.fit_transform(df[['Ecology_1', 'Ecology_2', 'Ecology_3']]), columns=df[['Ecology_1', 'Ecology_2', 'Ecology_3']].columns)
#         df['Ecology'] = eco_scaled['Ecology_1'] + eco_scaled['Ecology_2'] + eco_scaled['Ecology_3']
        
#         social_scaled = pd.DataFrame(scaler.fit_transform(df[['Social_1', 'Social_2', 'Social_3']]), columns=df[['Social_1', 'Social_2', 'Social_3']].columns)
#         df['Social'] = social_scaled['Social_1'] + social_scaled['Social_2'] + social_scaled['Social_3']
        
#         HC_scaled = pd.DataFrame(scaler.fit_transform(df[['Healthcare_1', 'Helthcare_2']]), columns=df[['Healthcare_1', 'Helthcare_2']].columns)
#         df['HC'] = HC_scaled['Healthcare_1'] + HC_scaled['Helthcare_2']
        
#         shops_scaled = pd.DataFrame(scaler.fit_transform(df[['Shops_1', 'Shops_2']]), columns=df[['Shops_1', 'Shops_2']].columns)
#         df['Shops'] = shops_scaled['Shops_1'] + shops_scaled['Shops_2']
        
#         df['Care'] = (df['Ecology'] + df['Social'] + df['HC'] + df['Shops']) / 4
        
#         df['SocialShops'] = df['Shops'] / df['Social']

        df['HouseYear'] = np.where((df['HouseYear'] <= 1910), 1, df['HouseYear'])
        df['HouseYear'] = np.where(((1910 < df['HouseYear']) & (df['HouseYear'] <= 1920)), 2, df['HouseYear'])
        df['HouseYear'] = np.where(((1920 < df['HouseYear']) & (df['HouseYear'] <= 1930)), 3, df['HouseYear'])
        df['HouseYear'] = np.where(((1930 < df['HouseYear']) & (df['HouseYear'] <= 1940)), 4, df['HouseYear'])
        df['HouseYear'] = np.where(((1940 < df['HouseYear']) & (df['HouseYear'] <= 1950)), 5, df['HouseYear'])
        df['HouseYear'] = np.where(((1950 < df['HouseYear']) & (df['HouseYear'] <= 1960)), 6, df['HouseYear'])
        df['HouseYear'] = np.where(((1960 < df['HouseYear']) & (df['HouseYear'] <= 1970)), 7, df['HouseYear'])
        df['HouseYear'] = np.where(((1970 < df['HouseYear']) & (df['HouseYear'] <= 1980)), 8, df['HouseYear'])
        df['HouseYear'] = np.where(((1980 < df['HouseYear']) & (df['HouseYear'] <= 1990)), 9, df['HouseYear'])
        df['HouseYear'] = np.where(((1990 < df['HouseYear']) & (df['HouseYear'] <= 2000)), 10, df['HouseYear'])
        df['HouseYear'] = np.where(((2000 < df['HouseYear']) & (df['HouseYear'] <= 2010)), 11, df['HouseYear'])
        df['HouseYear'] = np.where(((2010 < df['HouseYear']) & (df['HouseYear'] <= 2020)), 12, df['HouseYear'])
        
        return df
    
    

    def reduce_mem_usage(df):
        
        """ Оптимизируем размерности колонок, для того, чтобы они занимали меньше места в памяти. """

        start_mem = df.memory_usage().sum() / 1024**2
        print('Memory usage of dataframe is {:.2f} MB'.format(start_mem))

        for col in df.columns:
            col_type = df[col].dtype

            if col_type != object:
                c_min = df[col].min()
                c_max = df[col].max()
                if str(col_type)[:3] == 'int':
                    if c_min > np.iinfo(np.int8).min and c_max < np.iinfo(np.int8).max:
                        df[col] = df[col].astype(np.int8)
                    elif c_min > np.iinfo(np.int16).min and c_max < np.iinfo(np.int16).max:
                        df[col] = df[col].astype(np.int16)
                    elif c_min > np.iinfo(np.int32).min and c_max < np.iinfo(np.int32).max:
                        df[col] = df[col].astype(np.int32)
                    elif c_min > np.iinfo(np.int64).min and c_max < np.iinfo(np.int64).max:
                        df[col] = df[col].astype(np.int64)  
                else:
                    if c_min > np.finfo(np.float32).min and c_max < np.finfo(np.float32).max:
                        df[col] = df[col].astype(np.float32)
                    else:
                        df[col] = df[col].astype(np.float64)
            else:
                df[col] = df[col].astype('category')

        end_mem = df.memory_usage().sum() / 1024**2
        print('Memory usage after optimization is: {:.2f} MB'.format(end_mem))
        print('Decreased by {:.1f}%'.format(100 * (start_mem - end_mem) / start_mem))

        return df
```

```python
target_name = 'Price'
feature_names = ['DistrictId',
 'Square',
# 'KitchenSquare',
# 'HouseFloor',
 'HouseYear',
 'Ecology_1',
 'Ecology_2',
 'Ecology_3',
 'Social',
# 'Social_1',
# 'Social_2',
 'Social_3',
 'Healthcare_1',
 'Helthcare_2',
# 'Shops_1',
 'Shops_2',
 'RoomSquare',
 'FloorRank',
 'KitchenCoef']
```

```python
X = df[feature_names]
y = df[target_name]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
def evaluate_preds(true_values, pred_values, save=False):
    """Оценка качества модели и график preds vs true"""
    
    print("R2:\t" + str(round(r2(true_values, pred_values), 3)) + "\n" +
          "RMSE:\t" + str(round(np.sqrt(mse(true_values, pred_values)), 3)) + "\n" +
          "MSE:\t" + str(round(mse(true_values, pred_values), 3))
         )
    
    plt.figure(figsize=(8,8))
    
    sns.scatterplot(x=pred_values, y=true_values)
    plt.plot([0, 500000], [0, 500000], linestyle='--', color='black')  # диагональ, где true_values = pred_values
    
    plt.xlabel('Predicted values')
    plt.ylabel('True values')
    plt.title('True vs Predicted values')
    
    if save == True:
        plt.savefig(REPORTS_FILE_PATH + 'report.png')
    plt.show()
```

```python
gb_model = GradientBoostingRegressor(criterion='mse',
                                     max_depth=5,
                                     min_samples_leaf=9,
                                     random_state=42,  
                                     n_estimators=200)
gb_model.fit(X_train, y_train)

y_train_preds = gb_model.predict(X_train)
evaluate_preds(y_train, y_train_preds)
```

```python
y_predict = gb_model.predict(X_test)
evaluate_preds(y_test, y_predict)
```
